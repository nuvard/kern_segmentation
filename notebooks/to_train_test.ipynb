```python
import pandas as pd
import os
from PIL import Image
```

```python
data_uf = pd.read_csv('../data/data_uf.csv')
```

```python
data_dc = pd.read_csv('../data/data_dc.csv')
```

```python
i=0
bad_ids = []
for idx, data in data_uf.iterrows():
    
    dc_img_name = os.path.join('/project/data/',
                                data['Folder']+'/data/'+str(data['Id'])+'.jpeg')
        
    uf_img_name = os.path.join('/project/data/',
                                data['Folder']+'/data/'+str(data['Id'])+'.jpeg')
    layer_width = data["LayerDown"] - data["LayerTop"]    
    image_dc = Image.open(dc_img_name)
    crop_size = min(260,image_dc.size[1],image_dc.size[0], int(image_dc.size[1]*(0.2/layer_width))-1)
        #print(crop_size)
    print(image_dc.size[1]*0.2/layer_width,layer_width, image_dc.size[0])
        
    if (crop_size<260/5):
        i+=1
        print(crop_size, idx, dc_img_name)
        bad_ids.append(idx)
        
print( i)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test = train_test_split(data_uf, test_size=0.33, random_state=42)
```

```python
X_train.to_csv('../data/train_data_dc.csv', index=False)
X_test.to_csv('../data/test_data_dc.csv', index=False)
```

```python
ids = set(data_dc.index)-set(bad_ids)
```

```python
train_uf=data_uf.loc[X_train.index,]
test_uf=data_uf.loc[X_test.index,]
```
